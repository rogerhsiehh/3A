## Link other data

### Weather data

```python
from meteostat import Point, Hourly
from datetime import datetime, timedelta
```

```python
# Calculate the average latitude and longitude for the center
center_lat = trip['Pickup Centroid Latitude'].mean()
center_lon = trip['Pickup Centroid Longitude'].mean()
print(f"Center of pickup trip in Chicago (Average Coordinates): Latitude {center_lat}, Longitude {center_lon}")
```

```python
# Define the location using Point
location = Point(center_lat, center_lon)

# Define the date range for 2023
start = datetime(2023, 1, 1)
end = datetime(2023, 12, 31)

# Fetch hourly data
data = Hourly(location, start, end)
data = data.fetch()
data.reset_index(inplace=True)
```

```python
data.head()
```

```python
# Merge trip data with weather data on date and hour
trip['Date'] = trip['Trip Start Timestamp'].dt.date
trip['Hour'] = trip['Trip Start Timestamp'].dt.hour
data['Date'] = pd.to_datetime(data['time']).dt.date
data['Hour'] = pd.to_datetime(data['time']).dt.hour

trip = pd.merge(trip, data, on=['Date', 'Hour'], how='left')

# Print or further process the merged DataFrame
trip.drop(['time'], axis=1, inplace=True)
```

# Temporal Analysis

## Taxi demand

```python
df =trip.copy()
df['Hour'] = df['Trip Start Timestamp'].dt.hour
df['Day'] = df['Trip Start Timestamp'].dt.date
df['Weekday'] = df['Trip Start Timestamp'].dt.weekday
df['Month'] = df['Trip Start Timestamp'].dt.month
df['Weekofyear'] = df['Trip Start Timestamp'].dt.weekofyear

# Calculate traffic volumes
hourly_traffic = df.groupby('Hour').size().reset_index(name='Traffic Volume')
daily_traffic = df.groupby('Day').size().reset_index(name='Traffic Volume')
weekly_traffic = df.groupby('Weekday').size().reset_index(name='Traffic Volume')
monthly_traffic = df.groupby('Month').size().reset_index(name='Traffic Volume')
weekly_traffic = df.groupby('Weekofyear').size().reset_index(name='Traffic Volume')
```

### Daily traffic volume as line plot

```python
# Plot the daily traffic volume
plt.figure(figsize=(12, 6))
sns.lineplot(x='Day', y='Traffic Volume', data=daily_traffic, color="cornflowerblue")
plt.title('Daily Taxi Traffic Volume')
plt.ylabel('Traffic Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### By Hour

```python
# Plot the hourly traffic volume
plt.figure(figsize=(12, 6))
sns.barplot(x='Hour', y='Traffic Volume', data=hourly_traffic, color="cornflowerblue")
# Add title and labels
plt.title('Sum of Taxi Traffic Volume by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Traffic Volume')

# Show the plot
plt.show()
```

### By Weekday

```python
# Plot the weekly traffic volume
plt.figure(figsize=(12, 6))
sns.barplot(x='Weekday', y='Traffic Volume', data=weekly_traffic, color="cornflowerblue")
plt.title('Sum of Taxi Traffic Volume by Weekday')
plt.xlabel('Week')
plt.ylabel('Traffic Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### By Month

# Spatial Analysis

## Taxi Demand

### By Census Tract

census tract shapefile download from Chicago data portal https://data.cityofchicago.org/Facilities-Geographic-Boundaries/Boundaries-Census-Tracts-2010/5jrd-6zik

```python
# Aggregate traffic volume by Pickup Census Tract
pickup_traffic = df.groupby('Pickup Census Tract')['Trip Miles'].size().reset_index(name='Traffic Volume')
pickup_traffic["Log Daily Average"] = np.log(pickup_traffic["Traffic Volume"]/len(df["Date"].unique()))
```

```python
census_tract_shapefile = gpd.read_file('Boundaries - Census Tracts - 2010/geo_export_43205675-be18-433e-beb4-7eb585400a4e.shp')

# Convert both columns to string and strip any leading/trailing whitespace
pickup_traffic['Pickup Census Tract'] = pickup_traffic['Pickup Census Tract'].astype(str).str.strip()
pickup_traffic['Pickup Census Tract'] = pickup_traffic['Pickup Census Tract'].apply(lambda x: x.split('.')[0])
census_tract_shapefile['geoid10'] = census_tract_shapefile['geoid10'].astype(str).str.strip()

# Merge the traffic data with the geographic data
merged_gdf = census_tract_shapefile.merge(pickup_traffic, left_on='geoid10', right_on='Pickup Census Tract', how='left')

# Fill missing values in 'Traffic Volume' with 0
merged_gdf['Log Daily Average'] = merged_gdf['Log Daily Average'].fillna(0)

# Plot the data
plt.rcParams['figure.figsize'] = [10, 10]
fig, ax = plt.subplots()
merged_gdf.plot(ax=ax, column='Log Daily Average', cmap='OrRd')
census_tract_shapefile.plot(ax=ax, edgecolor='black', facecolor='none')
plt.title('Logarithm Daily Average Traffic Volume by Pickup Census Tract')
plt.show()
```

## Trip Length

### By Census tract

```python
# Calculate the average trip seconds by Pickup Census Tract
pickup_traffic = df.groupby('Pickup Census Tract')['Trip Seconds'].mean().reset_index(name='Average Trip Seconds')
```

```python
# Load the census tract shapefile
census_tract_shapefile = gpd.read_file('Boundaries - Census Tracts - 2010/geo_export_43205675-be18-433e-beb4-7eb585400a4e.shp')

# Convert both columns to string and strip any leading/trailing whitespace
pickup_traffic['Pickup Census Tract'] = pickup_traffic['Pickup Census Tract'].astype(str).str.strip()
pickup_traffic['Pickup Census Tract'] = pickup_traffic['Pickup Census Tract'].apply(lambda x: x.split('.')[0])
census_tract_shapefile['geoid10'] = census_tract_shapefile['geoid10'].astype(str).str.strip()

# Merge the traffic data with the geographic data
merged_gdf = census_tract_shapefile.merge(pickup_traffic, left_on='geoid10', right_on='Pickup Census Tract', how='left')

# Fill missing values in 'Log Average Trip Seconds' with 0
merged_gdf['Average Trip Seconds'] = merged_gdf['Average Trip Seconds'].fillna(0)

# Plot the data
plt.rcParams['figure.figsize'] = [10, 10]
fig, ax = plt.subplots()
merged_gdf.plot(ax=ax, column='Average Trip Seconds', cmap='OrRd', legend=True)
census_tract_shapefile.plot(ax=ax, edgecolor='black', facecolor='none')
plt.title('Average Trip Seconds by Pickup Census Tract')
plt.show()
```
